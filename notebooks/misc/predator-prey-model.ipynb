### Setting up the model

```python
alp0 = 0.05
alp1 = 0.3
alp2 = 0.02
alp3 = 0.05
alp4 = 0.06
alp5 = 0.01
alp6 = 0.03

#h = 0.05
h = 0.01

inits1 = [50,20,5,0]
#ends = [100,100,100,100]
#ends1 = [100,100,100,100]
ends1 = [1000,1000,1000,1000]

def fx1(curr_vars):
    curr_x = curr_vars[0]
    curr_y = curr_vars[1]
    curr_z = curr_vars[2]
    curr_t = curr_vars[3]
    return alp0*curr_x - alp1*curr_x*curr_y

def fy1(curr_vars):
    curr_x = curr_vars[0]
    curr_y = curr_vars[1]
    curr_z = curr_vars[2]
    curr_t = curr_vars[3]
    return alp3*curr_x*curr_y - alp4*curr_y*curr_z - alp2*curr_y

def fz1(curr_vars):
    curr_x = curr_vars[0]
    curr_y = curr_vars[1]
    curr_z = curr_vars[2]
    curr_t = curr_vars[3]
    return alp6*curr_y*curr_z - alp5*curr_z

def ft(curr_vars):
    curr_x = curr_vars[0]
    curr_y = curr_vars[1]
    curr_z = curr_vars[2]
    curr_t = curr_vars[3]
    return 1
```

```python
alp0 = 0.05
alp1 = 0.3
alp2 = 0.02
alp3 = 0.05
alp4 = 0.06
alp5 = 0.01
alp6 = 0.03

inits1 = [50,20,5,0]

def PPM_heuns():
    pts = dynamic_heuns(h, [[],[],[],[]], [fx1,fy1,fz1,ft],[inits1,ends1])
    print(len(pts))
    xs1 = [p[0] for p in pts]
    ys1 = [p[1] for p in pts]
    zs1 = [p[2] for p in pts]
    ts1 = [p[3] for p in pts]

    fig = figure(1)
    plt.title("Heun's approximation of Predator-Prey Model: x(t)")
    plot(ts1,xs1, label=r'$x_{t}$')
    plot(ts1,ys1, label=r'$y_{t}$')
    plot(ts1,zs1, label=r'$z_{t}$')
    plt.legend(loc='upper right')
    
PPM_heuns()
```
