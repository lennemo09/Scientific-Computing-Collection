## Model:
Given the system: 
$$x_{t+1} = mx_{t} + c$$


### Code for generating the plots

```python
def f(x, m ,c):
    return m*x + c

def xt(m,c, x0):
    xts = []
    xs = []
    x = x0
    t = 0
    while t < 15:
        xs.append(x)
        xt = f(x, m ,c)
        xts.append(xt)
        x = xt
        t += 1
    return xs, xts
```

### Plotting

```python
m = 0.8
c = 1
x0 = 0

xs,xts = xt(m,c,x0)

print(xts)
print(xs)

plot([min(xts+xs), max(xts)], [min(xts+xs), max(xts)], color="black", label='x=y')


cob_iterations = 15

last_x = x0
last_y = f(x0,m,c)
for _ in range(cob_iterations):
    next_x = f(last_x, m ,c)
    # Plot vertical line 
    plot([last_x, last_x], [last_y, next_x], color="orange")
    # Plot horizontal line
    plot([last_x, next_x], [next_x, next_x], color="green")

    last_x, last_y = next_x, next_x
    
plot(xs,xts, marker='o', label='xt')
plot([x0], [f(x0,m,c)], marker='x', markersize=10, mew=2, color="red",label='Start point')
plt.suptitle("Case 1 - A: m = {}, c = {}".format(m,c))
if m != 1:
    x_star = c/(1-m)
    plot([x_star], [f(x_star,m,c)], marker='+', markersize=10, mew=2, color="purple",label='Equilibrium')
plt.legend()
```
