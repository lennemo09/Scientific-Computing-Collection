## SEIR Model

```python
gamma = 0.2 # Transmission to exposed/infected rate
alpha = 0.01 # Exposed/Infected -> Recovered
epsilon = 0.002 # Exposed -> Infected
delta = 0.001 # Infected -> Death
f = 3*10**-5 # Transmission rate

def SEIR(vals,t=0):
    """
    Model:
    S -> E -> I -> R
    S -> I -> D
    E -> R
    
    dS/dt = -f*S*I
    dE/dt = gamma*f*S*I - alpha*E - epsilon*E
    dI/dt = (1-gamma)*f*S*I + epsilon*E - alpha*I - delta*I
    dR/dt =  alpha*I + alpha*E
    
    @param: vals: An array in the form [S(t), E(t), I(t), R(t)], that is:
                    vals[0] = S
                    vals[1] = E
                    vals[2] = I
                    vals[3] = R.
    """
    St = vals[0]
    Et = vals[1]
    It = vals[2]
    Rt = vals[3]
    
    dSdt = -f*St*It
    dEdt = gamma*f*St*It - alpha*Et - epsilon*Et
    dIdt = (1-gamma)*f*St*It + epsilon*Et - alpha*It - delta*It
    dRdt =  alpha*It + alpha*Et
    
    return array([dSdt, dEdt, dIdt, dRdt])
```

```python
SEIR_t = linspace(0,500,550)
SEIR_init = [1500, 20, 100, 0]

SEIR_sol = odeint(SEIR, SEIR_init, SEIR_t)
S, E, I, R = SEIR_sol.T
#print(S,E,I,R)

plot(SEIR_t, S, label='S')
plot(SEIR_t, E, label='E')
plot(SEIR_t, I, label='I')
plot(SEIR_t, R, label='R')
plt.grid()
plt.legend()

gamma = 0.2 # Transmission to exposed/infected rate
alpha = 0.01 # Exposed/Infected -> Recovered
epsilon = 0.002 # Exposed -> Infected
delta = 0.001 # Infected -> Death
f = 3*10**-5 # Transmission rate
```

## SEIRDS Model

```python
def SEIRDS(vals,t=0):
    """
    Model:
    S -> E -> I -> R -> S
    S -> I -> D
    E -> R
    
    dS/dt = -f*S*I
    dE/dt = gamma*f*S*I - alpha*E - epsilon*E
    dI/dt = (1-gamma)*f*S*I + epsilon*E - alpha*I - delta*I
    dR/dt =  alpha*I + alpha*E
    
    @param: vals: An array in the form [S(t), E(t), I(t), R(t)], that is:
                    vals[0] = S
                    vals[1] = E
                    vals[2] = I
                    vals[3] = R.
    """
    St = vals[0]
    Et = vals[1]
    It = vals[2]
    Rt = vals[3]
    
    dSdt = -f*St*It + gamma*rho*Rt 
    dEdt = gamma*f*St*It - alpha*Et - epsilon*Et
    dIdt = (1-gamma)*f*St*It + epsilon*Et - alpha*It - delta*It
    dRdt =  alpha*It + alpha*Et - gamma*rho*Rt - death*Rt
    
    return array([dSdt, dEdt, dIdt, dRdt])
```
